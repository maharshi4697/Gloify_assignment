```python
import pandas as pd
import json
import math
```

```python
# reading the data
url = 'https://raw.githubusercontent.com/rashida048/Datasets/master/movie_dataset.csv'
df = pd.read_csv(url,index_col=0)
```

```python
# keep only required columns
df = df[['budget', 'release_date', 'production_companies', 'popularity', 'revenue', 'runtime', 'title', 'vote_average']]

# filtering data to remove all movies that do not have a run-time.
filter = df['runtime']>0
filtered_df = df.where(filter).dropna()

# set default option
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Question 1: Sort the movies ascending and descending based on runtime.

asc_sorted_df = filtered_df.sort_values(by = 'runtime', ascending = True).reset_index(drop=True)
desc_sorted_df = filtered_df.sort_values(by = 'runtime', ascending = False).reset_index(drop=True)

#print('Movie with the shorted runtime: '+ asc_sorted_df['title'][0] + ' with a runtime of: ' + str(asc_sorted_df['runtime'][0]) + ' minutes')
#print('Movie with the longest runtime: '+ desc_sorted_df['title'][0] + ' with a runtime of: ' + str(desc_sorted_df['runtime'][0]) + ' minutes')
```

```python
df['production_company_names'] = ''
df['adjusted_budget'] = 0
df['adjusted_revenue'] = 0
for pos in range(len(df)):
    names = [company['name'] for company in json.loads(df['production_companies'][pos])]
    df['production_company_names'][pos] = names  
    df['adjusted_budget'][pos] = df['budget'][pos]/len(names)
    df['adjusted_revenue'][pos] = df['revenue'][pos]/len(names)
    
exploded_df = df.explode('production_company_names').reset_index(drop=True)

years = []
for row in exploded_df['release_date']:  
    if row:
        years.append(str(row).split('-')[0])
    
exploded_df['release_year'] = years
```

```python
grouped_production_budget = exploded_df.groupby('production_company_names')['adjusted_budget'].sum().reset_index()
top_5_prod_companies = grouped_production_budget.sort_values(by = 'adjusted_budget', ascending = False).reset_index(drop=True).head(5)
```

```python
#top_5_prod_companies
```

```python
top_5_vote_avg_agg_sorted = pd.DataFrame(columns=exploded_df.columns)

for company in top_5_prod_companies['production_company_names'].tolist():
    temp_df = exploded_df[exploded_df['production_company_names']==company]	
    temp_df = temp_df.sort_values(by = 'vote_average', ascending = False).head(5)
    top_5_vote_avg_agg_sorted = pd.concat([top_5_vote_avg_agg_sorted, temp_df], ignore_index=True)
```

```python
#top_5_vote_avg_agg_sorted[['title', 'production_company_names', 'vote_average']]
```

```python
top_5_popularity_agg_sorted = pd.DataFrame(columns=exploded_df.columns)

for company in top_5_prod_companies['production_company_names'].tolist():
    temp_df = exploded_df[exploded_df['production_company_names']==company]	
    temp_df = temp_df.sort_values(by = 'popularity', ascending = False).head(5)
    top_5_popularity_agg_sorted = pd.concat([top_5_popularity_agg_sorted, temp_df], ignore_index=True)
```

```python
top_5_revenue_agg_sorted = pd.DataFrame(columns=exploded_df.columns)

for company in top_5_prod_companies['production_company_names'].tolist():
    temp_df = exploded_df[exploded_df['production_company_names']==company]	
    temp_df = temp_df.sort_values(by = 'revenue', ascending = False).head(5)
    top_5_revenue_agg_sorted = pd.concat([top_5_revenue_agg_sorted, temp_df], ignore_index=True)
```

```python
year_sorted = [] 
for year in years: 
    if not math.isnan(float(year)):
        if year not in year_sorted and int(year) > 2000 and int(year) <= 2016: 
            year_sorted.append(year)           
year_sorted.sort()
```

```python
most_movies_by_year = pd.DataFrame(columns=['release_year', 'production_company_names',	'title'])
for year in year_sorted:
    temp_df = exploded_df[exploded_df['release_year']==year]	
    temp_df = temp_df.groupby(['release_year','production_company_names'])['title'].count().reset_index()
    temp_df = temp_df.sort_values(by = 'title', ascending = False).reset_index(drop=True).head(1)
    most_movies_by_year = pd.concat([most_movies_by_year, temp_df], ignore_index=True)
```

```python
sliced_df = exploded_df[exploded_df['production_company_names'].isin(['DC Comics','Marvel Studios'])] 
sliced_df = sliced_df.groupby('production_company_names')['adjusted_budget','adjusted_revenue'].sum().reset_index()
sliced_df['profit_ratio']=sliced_df['adjusted_revenue']/sliced_df['adjusted_budget']
```
